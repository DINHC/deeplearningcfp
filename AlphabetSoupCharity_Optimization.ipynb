```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
# Optimization1 = Dropping more columns 
```

```python
df = application_df.drop(['EIN','NAME','ORGANIZATION',"SPECIAL_CONSIDERATIONS"], axis=1)
df.head()
```

```python
df.shape
```

```python
df.nunique()
```

```python
df['APPLICATION_TYPE'].value_counts()
```

```python
application_types_to_replace = ('T9','T13','T12','T2','T25','T14','T29','T15','T17')

for app in application_types_to_replace:
    df['APPLICATION_TYPE'] =df['APPLICATION_TYPE'].replace(app,"Other")


df['APPLICATION_TYPE'].value_counts()
```

```python
df['CLASSIFICATION'].value_counts()
```

```python
df['CLASSIFICATION'].value_counts().loc[lambda x : x>1]
```

```python
classifications_to_replace = df['CLASSIFICATION'].value_counts().loc[lambda x : x < 900].index

for cls in classifications_to_replace:
    df['CLASSIFICATION'] = df['CLASSIFICATION'].replace(cls,"Other")

df['CLASSIFICATION'].value_counts()
```

```python
df.head()
```

```python
dummiesdf = pd.get_dummies(df, columns=['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE','INCOME_AMT'], drop_first=True, dtype=float)
dummiesdf.shape
```

```python
dummiesdfx = dummiesdf.drop('IS_SUCCESSFUL', axis=1)
dummiesdfx.head()
```

```python
dummiesdfy = dummiesdf['IS_SUCCESSFUL']
dummiesdfy
```

```python
X = dummiesdfx
y = dummiesdfy

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 50)
```

```python
scaler = StandardScaler()

X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
nn1 = tf.keras.models.Sequential()
#changed input_dim from 36 to 32
nn1.add(tf.keras.layers.Dense(units=50, activation="relu",input_dim=32))

nn1.add(tf.keras.layers.Dense(units=30, activation="relu"))

nn1.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

nn1.summary()
```

```python
nn1.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
import os
from tensorflow import keras
os.makedirs("checkpoints/", exist_ok=True)
chp = "checkpoints/weights.{epoch:05d}.hdf5"
```

```python
chp_callback = tf.keras.callbacks.ModelCheckpoint(filepath= chp, verbose =1, save_weights_only=True, save_freg = "epoch", period=5)
```

```python
fit_model_callback=nn1.fit(X_train_scaled, y_train, epochs = 100, callbacks=[chp_callback])
```

```python
model_loss, model_accuracy = nn1.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
#accuracy increased slightly with dropping of slightly more columns
```

```python
dummiesdf3 = pd.get_dummies(df3, columns=['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION','INCOME_AMT'], drop_first=True, dtype=float)
dummiesdf3.shape
```

```python
X3 = dummiesdfx3
y3 = dummiesdfy3

X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, random_state = 50)
```

```python
scaler = StandardScaler()

X3_scaler = scaler.fit(X3_train)

X3_train_scaled = X3_scaler.transform(X3_train)
X3_test_scaled = X3_scaler.transform(X3_test)
```

```python
def create_model(hp):
    nn3 = tf.keras.models.Sequential()

    activation = hp.Choice('activation',['relu','tanh','sigmoid'])

    nn3.add(tf.keras.layers.Dense(units=hp.Int('first_units',
            min_value=1,
            max_value=50,
            step=2
            ), activation=activation, input_dim=28))
    
    for i in range(hp.Int('num_layers', 1, 6)):
        nn3.add(tf.keras.layers.Dense(units=hp.Int('units_'+str(i),
            min_value=1,
            max_value=50,
            step=2),
            activation=activation))
    
    nn3.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    nn3.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])

    return nn3

```

```python
import kerastuner as kt
tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=20,
    hyperband_iterations=2)
```

```python
tuner.search(X3_train_scaled,y3_train,epochs=50,validation_data=(X3_test_scaled,y3_test))
```

```python
best_hyper = tuner.get_best_hyperparameters(1)[0]
best_hyper.values
```

```python
best_model = tuner.get_best_models(1)[0]
model_loss, model_accuracy = best_model.evaluate(X3_test_scaled,y3_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
